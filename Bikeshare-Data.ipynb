```python
# The libraries that we will use in the project.......
import time

import pandas as pd
import numpy as np
```

```python

cities=['chicago','new york city','washington']
months=['january', 'february', 'march', 'april', 'may', 'june','all']
days=['monday','tuesday','wednesday','thursday','friday','saturday','sunday','all']
```

```python
def get_filters():
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    # TO DO: get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    city=input('Select a city from Chicago, New York City and Washington. \n').lower()
    while city not in cities:
        city=input('Please write a valid city name from..... Chicago, New York City and Washington \n').lower()


    # TO DO: get user input for month (all, january, february, ... , june)
    
    month=input('enter month if you want to fillter month? if you want choose all month enter all\n').lower()
    while month not in months:
        month=input('Please write a valid month  name from.....all, january, february, ... , june \n').lower()
     
   # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)
    day=input('enter day if you want to fillter day? if you want choose all day enter all\n').lower()
    while day not in days:
        day=input('Please write a valid day name from.....all,monday, tuesday, ... sunday \n').lower()
    

    print('-'*40)
    return city, month, day
```

```python
def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
   
    # load data about city  into a data fram.....
    CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }
    
    df= pd.read_csv(CITY_DATA[city])              

    # convert Start and end  Time column to datetime by using to_datetime ....
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['End Time'] = pd.to_datetime(df['End Time'])
    # create new columns for month..day..start_end_combination..'start_end_difference....
    df['month'] = df['Start Time'].dt.month
    df['day'] = df['Start Time'].dt.day_name()
    df['start_end_combination'] = df['Start Station'].map(str) + ' to ' + df['End Station']
    df['start_end_difference'] = df['End Time'] - df['Start Time'] 
   
    #    # Determine the desired part of df from day....
    if month != 'all':
        month = months.index(month) + 1 #To match the month numbers in the month column ........ 
        df = df[df['month'] == month]

    # Determine the desired part of df from day....
    if day != 'all':
        df = df[df['day'] == day.title()]


    return df
```

```python
def time_stats(df):
    
    """Displays statistics on the most frequent times of travel."""
    
    
    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()
    
    # display the most common month
    month_mode_commen=df['month'].mode()[0]
    print('The most common month is: {}'.format(months[month_mode_commen-1]))
    
    # display the most common day of week
    most_common_day = df['day'].mode()[0]
    print('The most common day is: '+  most_common_day )
    
    # display the most common start hour
    df['hour of start'] = df['Start Time'].dt.hour
    most_common_hour = df['hour of start'].mode()[0]
    print('The most common start hour is: ' + str(most_common_hour))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def station_stats(df):
    """Displays statistics on the most popular stations and trip."""
    
    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()
    
    # display most commonly used start station
    most_commonly_used_start_station = df['Start Station'].mode()[0]
    print('The most common start station is: '+ most_commonly_used_start_station)
    
    # display most commonly used end station
    most_commonly_used_end_station = df['End Station'].mode()[0]
    print('The most common end station is: ' +   most_commonly_used_end_station )
    
    # display most frequent combination of start station and end station trip
    most_commonly_used_start_end_station = df['start_end_combination'].mode()[0]
    print('The most popular combination is: ' +  most_commonly_used_start_end_station )
        
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
       
```

```python
def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""
    
    print('\nCalculating Trip Duration...\n')
    start_time = time.time()
      
    # display total travel time
    total_travel_time = df['start_end_difference'].sum()
    hours =  total_travel_time.seconds // (60*60)
    minutes = total_travel_time.seconds % (60*60) // 60 
    seconds = total_travel_time.seconds % (60*60) % 60 
    print ('The total travel time is: '+ str(hours) + ' hours, ' + str(minutes) + 'minutes, ' + str(seconds) + 'seconds' )
    avarage_travel_time = df['start_end_difference'].mean()
    hours =  avarage_travel_time.seconds // (60*60)
    minutes = avarage_travel_time.seconds % (60*60) // 60 
    seconds = avarage_travel_time.seconds % (60*60) % 60 
    print ('The varage travel time is: '+ str(hours) + ' hours, ' + str(minutes) + 'minutes, ' + str(seconds) + 'seconds' )
    

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def user_stats(df):
    """Displays statistics on bikeshare users."""
    
    print('\nCalculating User Stats...\n')
    start_time = time.time()
    counts_of_user_types = df['User Type'].value_counts()
    # Display counts of user types
    print('counts of each user type ' + str(counts_of_user_types))
    
    # Display counts of gender
    if('Gender' not in df):
        print(' Gender data unavailable for Washington ')
    else:
        print('The genders are \n '+ str(df['Gender'].value_counts()))
    
    # Display earliest, most recent, and most common year of birth
    if ('Birth Year' not in df):
        print(' Birth year data unavailable for Washington')
    else:
        print('The Earliest birth year is: ' + str(df['Birth Year'].min()))
        print('The most recent birth year is:  '+ str(df['Birth Year'].max()))
        print('The most common birth year is:  ' + str(df['Birth Year'].mode()[0]))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def main(): 
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)
        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)
        view_datafrme(df)
        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
	main()
```
